```python
# === Actividad Semana 5 ===
# Manipulación e inspección de datos con pandas

import pandas as pd

# === Dataset base ===
datos = {
    'Nombre':  ['Ana', 'Bob', 'Clara', 'Diego', 'Eva'],
    'Edad':    [25, 30, 22, None, 28],
    'Ciudad':  ['Madrid', 'Lima', 'Bogotá', 'Medellín', None],
    'Ingreso': [3000, 4500, 2800, 5000, None]
}

df = pd.DataFrame(datos)

# Guarda el dataset base
df.to_csv('actividad_semana5.csv', index=False)
print("=== Dataset base ===")
print(df.head(), "\n")


# === 1) Series — crear y operar ===
serie_lista = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])
serie_dict = pd.Series({'x': 5, 'y': 10, 'z': 15})

# Acceder, modificar y operar
valor_b = serie_lista['b']
serie_lista['c'] = 35
serie_doble = serie_lista * 2

print("=== Series desde lista ===")
print(serie_lista)
print("\n=== Series desde diccionario ===")
print(serie_dict)
print("\n=== Series multiplicada por 2 ===")
print(serie_doble, "\n")


# === 2) DataFrame — crear y explorar ===
df2 = pd.DataFrame({
    'Producto': ['Mouse', 'Teclado', 'Monitor'],
    'Precio': [50, 80, 300],
    'Stock': [20, 10, 5]
})

print("=== DataFrame personalizado ===")
print(df2, "\n")

print("Tipos de columnas:")
print(df2.dtypes, "\n")


# === 3) Inspeccionar y explorar ===
print("=== Primeras filas (head) ===")
print(df.head(), "\n")

print("=== Últimas filas (tail) ===")
print(df.tail(), "\n")

print("=== Información general ===")
print(df.info(), "\n")

print("=== Forma del DataFrame ===")
print(df.shape, "\n")

print("=== Estadísticas básicas ===")
print(df.describe(), "\n")


# === 4) Acceder a columnas y filas ===
print("=== Columna 'Nombre' como Series ===")
print(df['Nombre'], "\n")

print("=== Fila con índice 1 usando loc ===")
print(df.loc[1], "\n")

print("=== Fila con índice 1 usando iloc ===")
print(df.iloc[1], "\n")

print("=== Valor específico (fila 1, columna 'Edad') ===")
print(df.loc[1, 'Edad'], "\n")


# === 5) Operaciones básicas ===
df['Edad'] = df['Edad'].fillna(0) + 5  # Maneja nulos sumando 5
df['Ingreso_anual'] = df['Ingreso'].fillna(0) * 12

print("=== DataFrame con operaciones ===")
print(df, "\n")


# === 6) Filtrar datos ===
mayores_30 = df[df['Edad'] > 30]
ciudades_filtradas = df[df['Ciudad'].isin(['Madrid', 'Lima'])]

print("=== Filas con Edad > 30 ===")
print(mayores_30, "\n")

print("=== Filas con Ciudad 'Madrid' o 'Lima' ===")
print(ciudades_filtradas, "\n")


# === 7) Manejo de valores faltantes ===
print("=== Conteo de valores faltantes ===")
print(df.isna().sum(), "\n")

df['Edad'].fillna(0, inplace=True)
df['Ciudad'].fillna('Desconocido', inplace=True)
df['Ingreso'].fillna(df['Ingreso'].median(), inplace=True)

print("=== DataFrame con nulos reemplazados ===")
print(df, "\n")


# === 8) Leer y guardar datos ===
df_leido = pd.read_csv('actividad_semana5.csv')
print("=== Archivo leído desde CSV ===")
print(df_leido.head(), "\n")

df_leido[['Nombre', 'Edad', 'Ciudad']].to_csv('actividad_semana5_reducido.csv', index=False)
print("Se ha guardado 'actividad_semana5_reducido.csv'\n")


# === 9) Ejercicio integrado ===
df_integrado = df.copy()
df_integrado['Edad'] = df_integrado['Edad'] + 5
df_integrado['Ingreso_anual'] = df_integrado['Ingreso'] * 12

df_filtrado = df_integrado[df_integrado['Ingreso_anual'] > 36000]
df_final = df_filtrado.sort_values(by='Ingreso_anual', ascending=False)

df_final.to_csv('personas_filtradas.csv', index=False)

print("=== Personas filtradas ===")
print(df_final)
```
